## 1 nuova colonna creata usando apply e lambda (commento).

ho creato una nuova colonna chiamata 'diff_2012_2016'. questa colonna contiene la differenze dei valori del 2016 e del 2012 per ogni paese. la colonna misura quanto è migliorato l'accesso all'elettricità in ciascun paese. valori positivi indicano un aumento della copertura e viceversa. questo valore è utile per identificare quali paesi hanno compiuto i maggiori progressi.

```python
df['diff_2012_2016']=df.apply(lambda x: x['2016']-x['2012'], axis=1)
```

```python
df.head(3)
```

se un paese partiva già con un livello alto di accesso all'elettricità è normale che la crescita non sia molto alta, per il smplice motivo che non c'è margine di crescita. confrontare la crescita assoluta tra il 2012 e il 2016 è fuorviante.

ho aggiunto una seconda colonna chiamate 'cresc_rel', che rappresenta la variazione percentuale dell'accesso all'elettricità rispetto al valore iniziale. in questo modo tengo conto che alcuni paesi molto elettrificati non possono crescere molto in termini assoluti. clacolando la crescita relativa si può identificare quali paesi hanno compiuto progressi in proporzione alla situazione di partenza.

```python
df['cresc_rel']=df.apply(lambda x:((x['2016']-x['2012'])/x['2012']*100) if x['2012']!=0 else None, axis=1)
```

```python
df.head(10)
```

se ci sono valori alti il paese è migliorato molto in proporzione, ovvero partiva da una situazione iniziale critica e ha fatto molti progressi. valori bassi significa che non c'era molto margine di miglioramento. valori negativi vuol dire che il paese non ha migliorato l'accesso, anzi è diminuito.

ho filtrato i peasi che avavamo una percentuale minore del 70% (paesi con una situazione svantaggiata). ho ordinato per crescita relativa in modo decrescente. in questo modo  posso individuare i paesi che sono migliorati di più rispetto al punto di partenza.

```python
s_critica=df[df['2012']<70]
s_critica_sort=s_critica.sort_values(by='cresc_rel', ascending=False)
s_critica_sort.head(10)
```

## 5 grafici comprensivi di titolo, labels e interpretazione (commento per ciascuno) di cui almeno 1 boxplot, 1 istogramma e 1 scatterplot usando sia Matplotlib che Seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

l'istogramma mostra come si distribuisce la crescita relativa dell'accesso all'elettricità tra il 2012 e il 2016. la maggior parte dei paesi presenta una crescita contenuta, altri paesi hanno valori più elevati, segno di forte miglioramento partendo da condizioni critiche. questo grafico serve per visualizzare la concentrazione dei valori e identificare eventuali outliers.

```python
plt.hist(df['cresc_rel'].dropna(), bins=20)
plt.title('distribuzione della crescita relativa tra il 2012 e il 2016')
plt.xlabel('crescita relativa')
plt.ylabel('numero paesi')
plt.show()
```

il grafico a barre mostra la media  dell'accesso all'elettricità nel 2016 suddiivo per continente. si può notare un divario energetico gloable: le regioni più sviluppate hanno raggiunto l'accesso universale all'elettricità, mentre le aree in via di sviluppo continuano a mostrare carenze signiifcative

```python
sns.barplot(data=df, x='Continent', y='2016')
plt.title('accesso alla corrente per continente')
plt.xlabel('continente')
plt.ylabel('accesso alla corrente')
plt.show()
```

questo boxplot mostra la distribuzioen dell'accccesos alla corrente nel 2016 suddiviso per continente. si può notare che l'Europa e il nord america hanno valori concentrati vicino al 100%. l'Africa  presenta una maggiore varibilità e i livelli più bassi. da questo grafico si possono evidenziare le disuguaglianze regionali globali.

```python
sns.boxplot(data=df, x='Continent', y='2016')
plt.title('distribuzione accesso corrente per continente anno 2016')
plt.xlabel('continente')
plt.ylabel('accesso alla corrente')
plt.show()
```

lo scatterplot mostra la relazione tra il livello iniziale de accesso all'elettricità nell 2012 e la crescita nel periodo successivo. si osserva che i paesi che partivano da una situazione critica tendono ad avere una crescita relativa più alta, mentre quelli già vicini all 100%mostrano variazioni minime. la suddivisione per continenti aiuta a visualizzare le differenze geografiche nei pettern di sviluppo.

```python
sns.scatterplot(data=df, x='2012', y='cresc_rel', hue='Continent')
plt.title('relazione  tra livello iniziale e crescia relativa')
plt.xlabel('accesso elettricità 2012')
plt.ylabel('crescita relativa')
plt.show()
```

il grafico mostra l'andamento medio dell'accesso all'elettricità dal 2012 al 2016 per ciascun continente. si mota una lenta,ma costante, crescita. l'africa mostra la traiettotoria dei crescitapiù marcata, ma parte da valori più bassi. il grafico evidenzia i progressi nel tempo e il persistere del divario tra regioni ad alto e basso sviluppo.

```python
df_media=df.groupby('Continent')[['2012', '2013', '2014', '2015', '2016']].mean()
df_media
```

```python
sns.lineplot(data=df_media.T)
plt.title('andamento medio accesso alla corrente dal 2012 al 2016')
plt.xlabel('anno')
plt.ylabel('accesso medio alla corrente (%)')
plt.legend(title='Continente')
plt.grid(True)
plt.show()
```
